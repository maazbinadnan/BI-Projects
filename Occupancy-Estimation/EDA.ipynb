```python
def detect_outliers_IQR(df):
   #Calculate the Q1:
   Q1 = np.percentile(df, 25)
   #Calculate the Q3:
   Q3 = np.percentile(df, 75)
   #Calculate the IQR:
   IQR=Q3-Q1
   # Upper bound
   upper = np.where(df >= (Q3+1.5*IQR))
   # Lower bound
   lower = np.where(df <= (Q1-1.5*IQR))
   #Outliers
   outliers = df[((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR)))]
   return outliers, upper, lower
```

```python
plt.figure(figsize=(10, 3), dpi = 100)
sns.boxplot(df['S5_CO2'])
```

```python
#find upper and lower quartile and percentage of outliers in s2_temp
outliers, upper, lower = detect_outliers_IQR(df['S5_CO2'])

print("number of outliers: "+ str(len(outliers)))

print("max outlier value: "+ str(outliers.max()))

print("min of outliers: "+ str(outliers.min()))

print("Percentage of outliers: "+ str(len(outliers)/len(df) * 100))
```

```python
extreme_temp = df[(df['S5_CO2'] > outliers.min()) & (df['S5_CO2'] < outliers.max())]
extreme_temp
```

```python
#PIE CHART of all temperature variables
labels = ['S1_Temp', 'S2_Temp', 'S3_Temp', 'S4_Temp']
sizes = [df['S1_Temp'].sum(), df['S2_Temp'].sum(), df['S3_Temp'].sum(), df['S4_Temp'].sum()]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.figure(figsize=(8,8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Temperature Variables')
plt.axis('equal')
plt.show()
```
