**ANOVA TEST**

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

y=df['use [kW]']
model = ols('y ~ C(Q("icon"))', data=df).fit()
anova_table=sm.stats.anova_lm(model, typ=2)
display(anova_table)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

y=df['use [kW]']
model = ols('y ~ C(Q("summary"))', data=df).fit()
anova_table=sm.stats.anova_lm(model, typ=2)
display(anova_table)
```

statistically significant as it confirms that the overall summary has an impact on the overall generation
